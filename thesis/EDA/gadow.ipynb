```python
import pandas as pd
from tqdm import tqdm
```

```python
data = pd.read_excel("/home/agata/urban-mobility-project/data/raw/KBR_Gadow_2019_bez_hasla.xlsx", engine='openpyxl', header=[0,1])
```

```python
l = list(data.columns)[199:]
l.append(list(data.columns)[0])
df = data[l]
```

```python
df.columns = list(map(lambda x: x[0], list(df.columns)))
df = df.rename(columns={'Unnamed: 0_level_0': 'ID odpowiedzi'})
```

```python
new_columns = [
    'user_id',
    'travel_id',
    'start_date',
    'duration_time',
    'mean_of_transport',
    'start_place_type',
    'finish_place_type',
    'start_place',
    'start_region',
    'residence_region',
    'finish_place'
]

travels_df = pd.DataFrame(columns=new_columns)

for i in tqdm(range(1, 28)):
    columns = [
        'ID odpowiedzi',
        f'ID podróży ({i})',
        f'Data rozpoczecia podróży ({i})',
        f'Czas trwania podróży [s] ({i})',
        f'Środek transportu ({i})',
        f'Typ miejsca startu ({i})',
        f'Typ miejsca zakończenia ({i})',
        f'Miejsce startu ({i})',
        f'Numer regionu miejsca startu ({i})',
        f'Numer regionu miejsca zakończenia ({i})',
        f'Miejsce zakończenia ({i})'
    ]

    sub_df = df[columns]
    sub_df = sub_df.dropna()
    sub_df.columns = new_columns

    travels_df = travels_df.append(sub_df)

travels_df['start_date'] = pd.to_datetime(travels_df['start_date'])
```

```python
tdf = pd.DataFrame(columns=['user_id', 'travel_id', 'duration_time', 'mean_of_transport', 'distance'])

for row in travels_df.itertuples():
    g_start = geocoder.osm(row[8])
    g_finish = geocoder.osm(row[11])

    try:
        start_coords = (g_start.osm['x'], g_start.osm['y'])
        end_coords = (g_finish.osm['x'], g_finish.osm['y'])

        tdf = tdf.append({
            'user_id': row[1],
            'travel_id': row[2],
            'distance': round(geocoder.distance(start_coords, end_coords, units='meters')),
            'duration_time': row[4],
            'mean_of_transport': row[5]
        }, ignore_index=True)

    except Exception as e:
        print()
        print(e)
        print(row[8])
        print(g_start.osm)
        print(row[11])
        print(g_finish.osm)

tdf
```
